```python
%matplotlib inline
import numpy as np
from scipy.integrate import quad
from scipy.integrate import fixed_quad
from math import *
from scipy.special import *
import matplotlib.pyplot as plt
def integr(t,omega):
    D = np.zeros(len(t))
    for i in range(len(t)):
        D[i] = cos((omega**2-1-2/j12)*t[i])*jv(0,t[i])*jv(0,t[i]/j12)**2
    return D
def numerical(omega):
    integ = fixed_quad(integr,-1000,1000,args=(omega,), n = 3000)
    return 2/3/pi*omega*integ[0]
def integr2(t,omega):
    D = np.zeros(len(t))
    for i  in range(len(t)):
        D[i] = cos((omega**2-1-1/j12)*t[i])*jv(0,t[i])*jv(0,t[i]/j12)
    return D
def numerical2(omega):
    integ = fixed_quad(integr2,-1000,1000,args=(omega,), n = 3000)
    return omega*integ[0]
def final_solving(j12, N):
    omega_deb = sqrt(2+2/j12)
    D_num = np.zeros(N); D_deb = np.zeros(N); D_num2 = np.zeros(N); D_deb2 = np.zeros(N); om_sqr = np.linspace(0,1.6,N)
    for i in range(len(om_sqr)):
        D_num[i] = numerical(om_sqr[i]); D_deb[i] = 3*om_sqr[i]**2/omega_deb**3
        D_num2[i] = numerical2(om_sqr[i]); D_deb2[i] = 2*om_sqr[i]/omega_deb**2
    norm = 0; norm2 = 0; delta = om_sqr[1] - om_sqr[0]
    for i in range(N-1):
        norm += (D_num[i+1]+D_num[i])*delta/2; norm2 += (D_num2[i+1]+D_num2[i])*delta/2
    D_num = D_num/norm; D_num2 = D_num2/norm2
    return om_sqr, D_num, D_deb, D_num2, D_deb2
```

```python
j12 = 2
N = 100
solution = final_solving(j12, N)
```

```python
plt.figure(figsize=(10,10))
plt.title('Трехмерная задача, j1/j2= '+ str(j12))
plt.plot(solution[0], solution[1], label = " Численное решение")
plt.plot(solution[0], solution[2], "r", label = 'Аналитическое приближение')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.1)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.title('Двумерная задача, j1/j2= '+ str(j12))
plt.plot(solution[0], solution[3], label = " Численное")
plt.plot(solution[0], solution[4],"r", label = "Аналитическое приближение")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.grid(True)
plt.show()
```
